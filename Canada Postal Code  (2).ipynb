### Map of Toronto Using Longitude and Latitude Values

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# Adding markers to the map
for lat, lng, borough, Neighbourhood in zip(canada_postalcodecombine['Latitude'], canada_postalcodecombine['Longitude'], canada_postalcodecombine['Borough'], canada_postalcodecombine['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
